```python
import os

# Change the current working directory to the specified location

# Get a list of all folders in the current working directory
folders = [folder for folder in os.listdir() if os.path.isdir(folder)]

# Iterate through each folder
for folder in folders:
    # Change the current working directory to the current folder
    os.chdir("/content/drive/MyDrive/Mp3-to-Midi/archive.zip (Unzipped Files)")
    os.chdir(folder)

    # Get a list of all files and folders in the current folder
    items = os.listdir()
    # Print the current folder name
    print(f"Current folder: {folder}")
    # Print each item in the current folder
    for item in items:
        #print(f"/content/drive/MyDrive/Mp3 to Midi/archive.zip (Unzipped Files)/{folder}/{item}")

        if(os.path.isfile(item[:-3] + "mp3") == False):
          print("   " + item)
          #fs.play_midi(item)
          fs.midi_to_audio(item, item[:-3] + "wav")
          AudioSegment.from_file(item[:-3] + "wav").export(item[:-3] + "mp3", format="mp3", bitrate="128k")
          os.remove(item[:-3] + "wav")
        else:
          print("     skipping")
    # Change back to the parent directory
    os.chdir("/")
```
