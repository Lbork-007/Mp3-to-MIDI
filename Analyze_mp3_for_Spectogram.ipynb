```python
import os
import matplotlib.pyplot as plt
import gc

#for loading and visualizing audio files
import librosa
import librosa.display

#to play audio
import IPython.display as ipd
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Get a list of all folders in the current working directory
folders = [folder for folder in os.listdir("/content/drive/MyDrive/Mp3-to-Midi/archive.zip (Unzipped Files)/")]

# Iterate through each folder
for folder in folders:
    # Change the current working directory to the current folder
        audio_fpath = "/content/drive/MyDrive/Mp3-to-Midi/archive.zip (Unzipped Files)/" + folder + "/"

        audio_clips = os.listdir(audio_fpath)

        # Print the current folder name
        print(f"Current folder: {folder}")
        for audio_clip in audio_clips:
            if audio_clip.endswith(".mp3"):
                print(f"   Current file: {audio_clip}")
                if not os.path.isfile(audio_fpath + audio_clip[:-4] + '.png'):
                    x, sr = librosa.load(audio_fpath+audio_clip, sr=44100)

                    # Only one figure
                    plt.figure(figsize=(14, 5))
                    X = librosa.stft(x)
                    Xdb = librosa.amplitude_to_db(abs(X))

                    # Display and save the spectrogram
                    librosa.display.specshow(Xdb, sr=sr)
                    plt.savefig(os.path.join(audio_fpath, audio_clip[:-4] + '.png'))
                    print("       saved")

                    # Clean up
                    plt.clf()
                    plt.close('all')  # Close all figures
                    del X, Xdb, x  # Free variables
                    gc.collect()  # Run garbage collection
                else:
                    print("      skipping")
```
