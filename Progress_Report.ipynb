```python
import os
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
folders = [folder for folder in os.listdir("/content/drive/MyDrive/Mp3-to-Midi/archive.zip (Unzipped Files)/")]
numOfUnfinishedMP3 = 0
numOffinishedMP3 = 0
numOfUnfinishedPNG = 0
numOffinishedPNG = 0
for folder in folders:
  audio_fpath = "/content/drive/MyDrive/Mp3-to-Midi/archive.zip (Unzipped Files)/" + folder + "/"
  audio_clips = os.listdir(audio_fpath)
  # Print the current folder name
  print(f"Current folder: {folder}")
  for audio_clip in audio_clips:
    if audio_clip.endswith(".mid"):
      if os.path.isfile(audio_fpath + audio_clip[:-4] + '.mp3') == False:
        numOfUnfinishedMP3 += 1
      else:
        numOffinishedMP3 += 1
      if os.path.isfile(audio_fpath + audio_clip[:-4] + '.png') == False:
        numOfUnfinishedPNG += 1
      else:
        numOffinishedPNG += 1
print(f"Number of unfinished Mp3s: {numOfUnfinishedMP3}")
print(f"Number of finished Mp3s: {numOffinishedMP3}")
print(f"Number of unfinished PNGs: {numOfUnfinishedPNG}")
print(f"Number of finished PNGs: {numOffinishedPNG}")
```
